```python
import pandas as pd
df=pd.read_csv("chapter3.csv")
```

```python
avg_price=df.groupby("Area")["Price"].mean()
print(avg_price)
```

```python
avg_price.sort_values(ascending=False)
```

```python
df.groupby("Area")["Price"].count().sort_values(ascending=False)
```

```python
df.groupby("Area").agg({"Price":"mean", "Bedrooms":"mean"})
```

```python
df.loc[df.groupby("Area")["Price"].idxmax()]
```

```python
df.loc[df.groupby("Area")["Price"].idxmin()]
```

```python
df.groupby("Area")["Price"].rank(ascending=False)
```

```python
df.groupby("Area").apply(lambda x: x.nlargest(3, "Price"))
```

```python
df[["Price", "Bedrooms", "Bathrooms"]].corr()
```

```python
df.nlargest(10, "Price")
```

```python
df.nsmallest(10, "Price")
```

```python
df["Price_Category"] = pd.cut(
    df["Price"],
    bins=3,
    labels=["Low", "Medium", "High"]
)
print(df["Price_Category"])
```

```python
Q1 = df["Price"].quantile(0.25)
Q3 = df["Price"].quantile(0.75)
IQR = Q3 - Q1

outliers = df[(df["Price"] < Q1 - 1.5*IQR) | 
              (df["Price"] > Q3 + 1.5*IQR)]
print(len(outliers))
```
