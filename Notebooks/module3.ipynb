# finding house age in years (new column)

```python
df["house_age"] = df["yr_built"].apply(lambda x: 2025 - x)
df[["yr_built", "house_age"]].head()
```

## creating price categories for better understanding of the house prices

```python
df["price_category"] = df["price"].apply(
    lambda x: "Low" if x < 300000 else ("Medium" if x < 800000 else "High")
)
df["price_category"].value_counts()
```

## Detect Outliers Flag using Lambda

```python
q1 = df["price"].quantile(0.25)
q3 = df["price"].quantile(0.75)
iqr = q3 - q1

df["price_outlier"] = df["price"].apply(
    lambda x: 1 if (x < (q1 - 1.5*iqr)) or (x > (q3 + 1.5*iqr)) else 0
)
df["price_outlier"].value_counts()
```

## Transform Skewed Columns using Lambda (ML-Ready EDA)


```python
df["log_price"] = df["price"].apply(lambda x: np.log1p(x))
```
