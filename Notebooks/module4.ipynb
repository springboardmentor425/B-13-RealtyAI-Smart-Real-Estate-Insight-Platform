```python
import pandas as pd
import numpy as np
df = pd.read_csv("D:/ISB/datasets/data_house/data_house.csv",index_col=0)
df.head()
```

## Average price per bedroom count

```python
bedroom_price = df.groupby("bedrooms")["price"].mean().sort_values(ascending=False)
bedroom_price
```

## Group by Year Built (Market Trend Analysis)

```python
year_price = df.groupby("yr_built")["price"].mean().sort_values(ascending=False)
year_price.head(10)
```

## Group Multiple Columns (Advanced EDA – High Impact)

```python
multi_group = df.groupby(["bedrooms", "bathrooms"])["price"].mean().sort_values(ascending=False)
multi_group.head(10)
```

### Cheapest houses

```python
df.sort_values(by="price", ascending=True).head(5)
```

### Top expensive houses

```python
df.sort_values(by="price", ascending=False).head(5)
```

## creating price categories for better understanding of the house prices

```python
df["price_category"] = df["price"].apply(
    lambda x: "Low" if x < 300000 else ("Medium" if x < 800000 else "High")
)
df["price_category"].value_counts()
```
