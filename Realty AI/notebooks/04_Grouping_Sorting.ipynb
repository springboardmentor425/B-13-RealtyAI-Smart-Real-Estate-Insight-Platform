## Groupwise analysis

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:/Users/dindu/OneDrive/Desktop/Realty AI/dataset/real_estate_dataset.csv")
```

```python
df.head()
```

```python
df.groupby("city").price.count()
```

```python
df.groupby("city").price.min()
```

```python
df.groupby("beds").apply(lambda x: x.city.iloc[0])
```

```python
df.groupby("city").apply(lambda x: x.price.iloc[0])
```

```python
df.groupby(["city", "beds"]).apply(lambda x: x.loc[x.price.idxmax()])
```

```python
df.groupby("city").price.agg([len, min, max])
```

## Multi-indexes

```python
city_beds = df.groupby(["city", "type"]).price.agg([len])
city_beds
```

```python
mi = city_beds.index
type(mi)
```

```python
city_beds.reset_index()
```

## Sorting

```python
city_beds = city_beds.reset_index()
city_beds
```

```python
city_beds.sort_values(by="len", ascending=False)
```

```python
city_beds.sort_index()
```

```python
city_beds.sort_values(by=["city", "len"])
```
