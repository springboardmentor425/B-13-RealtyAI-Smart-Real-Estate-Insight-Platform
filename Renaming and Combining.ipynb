```python
import pandas as pd

file_path = r"C:\Users\HARSHITHA\Downloads\price_paid_records.csv.zip"

df = pd.read_csv(
    file_path,
    compression="zip",
    usecols=[
        "Price",
        "Date of Transfer",
        "Property Type",
        "Old/New",
        "Duration",
        "Town/City",
        "District",
        "County"
    ]
)

df.head()
```

```python
df.columns
```

```python
df = df.rename(columns={"Date of Transfer": "Date"})
```

```python
df.columns
```

```python
# Convert to datetime first
df["Date"] = pd.to_datetime(df["Date"])

# Create new columns
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month

# Check columns
df.columns
```

```python
df = df.drop("Date", axis=1)
```

```python
city_counts = df["Town/City"].value_counts()
top10 = city_counts.head(10).index
df["Town/City"] = df["Town/City"].apply(lambda x: x if x in top10 else "Other")
```

```python
df["Town/City"].value_counts()
```

```python
df["Age_Duration"] = df["Old/New"] + "_" + df["Duration"]
```

```python
df["Age_Duration"].head()
```

```python
df["Property Type"].head()
```

```python
df["Property Type"] = df["Property Type"].replace({
    "D": "Detached",
    "S": "Semi-Detached",
    "T": "Terraced",
    "F": "Flat",
    "O": "Other"
})
```

```python
df["Property Type"].head()
```

```python
county_counts = df["County"].value_counts()
top_counties = county_counts.head(10).index

df["County"] = df["County"].apply(lambda x: x if x in top_counties else "Other")
```
