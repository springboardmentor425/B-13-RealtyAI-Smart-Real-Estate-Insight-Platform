```python
import matplotlib.pyplot as plt

avg_price = df.groupby("Area")["Price"].mean().sort_values(ascending=False)

avg_price.plot(kind="bar")
plt.title("Average Price by Area")
plt.xlabel("Area")
plt.ylabel("Average Price (RM)")
plt.xticks(rotation=90)
plt.show()
```
