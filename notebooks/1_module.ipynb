## Important Notes
When saving a dataframe to CSV using `to_csv("...")`, the resulting file includes an unnamed index column with increasing integers. To use this column as the index when reading the file back, use the `index_col` parameter in `read_csv()`. This ensures proper index restoration and cleaner data representation.
