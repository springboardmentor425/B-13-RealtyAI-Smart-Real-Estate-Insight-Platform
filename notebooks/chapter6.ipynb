```python
import pandas as pd
df=pd.read_csv("chapter5.csv")
```

```python
df.rename(columns={
    "Price": "Total_Price",
    "Area": "Location",
    "Category": "Property_Type"
})
```

```python
df.rename_axis("S_No.", axis="rows", inplace=True)
```

```python
df.head()
```

```python
df = df.drop(columns=["Unnamed: 0"])
df.head()
```

```python
df.columns = df.columns.str.replace("_", " ")
df.head()
```

```python
df.head()
```

```python
df["Description"] = df["Title"] + " - " + df["Area"]
df["Description"].head()
```

```python
df["Total Rooms"] = df["Bedrooms"] + df["Bathrooms"]
df["Total Rooms"].head()
```

```python
df.head()
```

```python
df["Fully_Furnished"] = df["Title"].str.contains("Fully Furnished", case=False, na=False)
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.heatmap(
    df[["Price", "Size", "Bedrooms", "Bathrooms", "Price per sqft"]].corr(),
    annot=True,
    cmap="coolwarm"
)
plt.title("Correlation Matrix")
plt.show()
```
