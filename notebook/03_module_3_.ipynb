#### Summary Functions & Grouping

```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/NYC_Rolling_Sales_Dataset/nyc-rolling-sales.csv')
```

##### Convert Text to Numeric

```python
df['SALE PRICE'] = pd.to_numeric(df['SALE PRICE'], errors='coerce')
```

##### Describe Numerical Columns

```python
df['SALE PRICE'].describe()
```

##### Group wise Summary

```python
# Convert BOTH columns to numbers properly
df['SALE PRICE'] = pd.to_numeric(df['SALE PRICE'], errors='coerce')
df['LAND SQUARE FEET'] = pd.to_numeric(df['LAND SQUARE FEET'], errors='coerce')
df['RESIDENTIAL UNITS'] = pd.to_numeric(df['RESIDENTIAL UNITS'], errors='coerce')

# Remove rows with any missing values
df_clean = df.dropna(subset=['SALE PRICE', 'LAND SQUARE FEET', 'RESIDENTIAL UNITS'])

# Now groupby will work!
result = df_clean.groupby('RESIDENTIAL UNITS')[['SALE PRICE', 'LAND SQUARE FEET']].mean().round(2)
print(result.head(10))
```

##### Categorize Sale Price

```python
def price_category(price):
    if price > 1_000_000:
        return 'High'
    elif price > 500_000:
        return 'Medium'
    else:
        return 'Low'

df['PRICE_CATEGORY'] = df['SALE PRICE'].apply(price_category)
df[['SALE PRICE', 'PRICE_CATEGORY']].head()
```

##### Combined Insight

```python
df.groupby('PRICE_CATEGORY')['SALE PRICE'].count()
```
