#### Indexing, Selecting & Assigning

##### Import Libraries

```python
import pandas as pd
import numpy as np
```

##### Data Loading

```python
df = pd.read_csv('../data/NYC_Rolling_Sales_Dataset/nyc-rolling-sales.csv')
```

##### Accessing the LAND SQUARE FEET Column

```python
land_sqft = df['LAND SQUARE FEET']
print(land_sqft.head(10))
```

##### Index-based Selection

```python
zip_codes_iloc = df.iloc[:, 11] # all rows and 11th column
print(zip_codes_iloc.head(10))
```

##### Select specific cell

```python
price_first = df.loc[0, 'SALE PRICE']
print(price_first)
```
